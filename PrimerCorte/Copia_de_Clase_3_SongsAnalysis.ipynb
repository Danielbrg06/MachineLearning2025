1.1 Cuanto es el promedio de la popularidad

```python
data['popularity'].mean()
```

2.Cual es el artista más popular

```python
data.groupby("artist")["popularity"].mean().sort_values(ascending=False)
```

3. Cual es el genero con mayor popularidad y cual es el genero mas popular

```python
data.groupby("genre")["popularity"].mean().sort_values(ascending=False)
```

4. Cual es el artista que tiene mas canciones con popularidades arriba de 70

```python
data[data['popularity']>70].groupby("artist")["popularity"].count().sort_values(ascending=False)
```

5. La cancion mas popular

```python
data.groupby("song")["popularity"].max().sort_values(ascending=False)
```

```python
data[data['popularity'] == data['popularity'].max()]
```
